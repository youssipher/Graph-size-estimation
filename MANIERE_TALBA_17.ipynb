Etant donné un graphe $G = (V , E)$ fini, non orienté et connexe, on cherche à estimer sa taille. On
suppose qu’on a accès au graphe de la manière suivante: on peut accéder à un sommet particulier
$i_0 \in V$ du graphe, et ayant accédé à un sommet $i$ du graphe, on peut alors accéder à n’importe quel
voisin de $i$ dans $G$. On va alors estimer $N$ par $\hat{N}=\frac{C_l^2}{2l}$ et étudier le comportement de cet estimateur pour différents types de graphes.

## Partie théorique
Le but de cette partie est d'esquisser quelques unes des propriétés de l'estimateur utilisé, avant d'effectuer des simulations.

### T.1 Expression de la loi de  probabilité de $Y_t$
Sans nuire à la généralité, nous avons considéré que (quitte à reordonner les sommets), nous partons de $i_0=1$ pour effectuer les marches aléatoires.
$A$ étant la matrice d'adjacence du graphe $G$, pour tout $\tau \in \mathbb{N}$, $A^\tau_ {1,i}$ donne le nombre de chemins de longueur $\tau$ entre le sommet $1$ fixé et le sommet $i$. $\\$ Ainsi les $X^\tau_i$ étant choisis uniformément, la probabilité d'aboutir en $i$ est égal au nombre de chemins entre $1$ et $i$, divisé par le nombre de chemins partant de $1$:
\begin{equation}\label{eq:}
\mathbb{P}(Y_{t} = i) = \frac{(A^\tau)_{1,i}}{d^\tau} 
\end{equation}

On obtient donc $\mathbb{P}(Y_{t}=i) = (P^{\tau})_{1,i}$

### T.2 Expression asymptotique de $\mathbb{P}$
$P$ étant symétrique, il existe, d'après le théorème spectral, $\Omega$ orthogonale telle que
\begin{align*}
P = \Omega \begin{bmatrix} 
1 & 0 & ... & 0 \\ 
0 & \lambda_2 & ...&0 \\ 
. & . & ...&. \\
0 & 0 & ... & \lambda_N
\end{bmatrix} \Omega^T 
\Rightarrow P^\tau &= \Omega \begin{bmatrix} 
1 & 0 & ... & 0 \\ 
0 & \lambda_2 ^\tau & ...&0 \\ 
. & . & .&. \\
0 & 0 & ... & \lambda_N^\tau
\end{bmatrix} \Omega^T\\
~Ainsi~ \lim_{\tau\to+\infty} P^\tau &= \Omega \lim_{\tau\to+\infty} \begin{bmatrix} 
1 & 0 & ... & 0 \\ 
0 & \lambda_2 ^\tau & ...&0 \\ 
. & . & .&. \\
0 & 0 & ... & \lambda_N^\tau
\end{bmatrix} \Omega^T  \\
\end{align*}
Puisque $\forall ~ i~ \in [2;N] ~~ |\lambda_i|< 1$, on obtient finalement:
$$\lim_{\tau\to+\infty} P^\tau = \Omega  \begin{bmatrix} 
1 & 0 & ... & 0 \\ 
0 & 0 & ...&0 \\ 
. & . & .&. \\
0 & 0 & ... & 0
\end{bmatrix}\Omega^T \nonumber$$
On peut alors poser $\lim_{\tau\to\infty} P^\tau = L$ et $M = \begin{bmatrix} 
1 & 0 & ... & 0 \\ 
0 & 0 & ...&0 \\ 
. & . & .&. \\
0 & 0 & ... & 0
\end{bmatrix}$
$\\$Puisque $M$ est de rang $1$, $L$ sera aussi de rang $1$ (et symétrique).
On peut poser $L = \begin{bmatrix} 
\alpha_1 & \beta_1\alpha_1 &\beta_2\alpha_1 &... & \beta_{N-1}\alpha_1 \\ 
\alpha_2 & \beta_1\alpha_2 & \beta_2\alpha_2&...& \beta_{N-1}\alpha_2 \\ 
        . & . & .& .&. \\
\alpha_N & \beta_1\alpha_N &\beta_2\alpha_N &... & \beta_{N-1}\alpha_N
\end{bmatrix}$

En notant $s_i : M \mapsto  \qquad \sum_{k=1}^{N} M_{i,k}$ la forme linéaire qui à une matrice associe la somme de ses coefficients de ligne $i$, la continuité de $s_i$ est assurée car $M_N(\mathbb{R})$ est de dimension finie.$\\$
Ainsi $\forall ~ i ~ \in ~ [1;n], s_i(P^\tau) \rightarrow s_i(L)$ lorsque $\tau \rightarrow +\infty$ mais on a aussi $s_i(P^\tau) = 1$ en tant que somme des probabilités d'atteindre un sommet $k$ en partant de $i$. Cela nous permet d'avoir $s_i(L)=1$

On obtient alors $\forall~ i \in~ [1,n]~~ \alpha_i(\sum_{k=1}^{N-1} \beta_k) = 1~~ \Rightarrow~~ \alpha_1 = \alpha_2=...=\alpha_N$

$L$ étant symétrique, on a aussi $\beta_1 =\beta_2=...=\beta_{N-1} = 1$

On obtient finalement $\sum_{k=1}^{N} \alpha_k = 1 ~~ \Rightarrow~~ \alpha_1 = \alpha_2=...=\alpha_N= \frac{1}{N}$

Donc $L = \begin{bmatrix} 
\frac{1}{N} & \frac{1}{N} & ... & \frac{1}{N} \\ 
\frac{1}{N} & \frac{1}{N} & ...&\frac{1}{N} \\ 
. & . & .&. \\
\frac{1}{N} & \frac{1}{N} & ... & \frac{1}{N}
\end{bmatrix}$

Et on tire $\mathbb{P}(Y_t=i)\rightarrow \frac{1}{N}$ lorsque $\tau \rightarrow +\infty$


$\mathbb{P}$ tend donc vers une loi uniforme sur l'ensemble des sommets, et cela peu importe la norme (dimension finie).

### T.3 
Lorsqu'on a $C_{l-1} = m$ , il y a eu $l-1$ collisions, et on a donc eu $m-(l-1)$ sommets différents comme sommet final d'une des marches aléatoires effectuées.

$\mathbb{P}(C_l - C_{l-1}>n|C_{l-1}=m)$ est la probabilité de ne pas avoir comme sommet final l'un de ceux déjà parcourus,et cela pour $n$ marches aléatoires successives. Les $Y_t$ étant uniformément distribués, cela revient à tirer $n$ sommets parmi les $N-m+l-1$ restants, sans remise.


Ainsi \begin{align*}\mathbb{P}(C_l - C_{l-1}>n|C_{l-1}=m) &= \frac{A^n_{N-m+l-1}}{N^n}\\  ~Ainsi~~~ \mathbb{P}(C_l - C_{l-1}>n|C_{l-1}=m) &= \frac{(N-m+l-1)(N-m+l-2)...(N-m+l-n)}{N^n} \end{align*}

### T.4 Qualité de l'estimateur
On a, d'après le point précédent, \begin{align*}\mathbb{P}(C_l - C_{l-1}>n|C_{l-1}=m) &= \frac{(N-m+l-1)(N-m+l-2)...(N-m+l-n)}{N^n} \\ &= \prod_{i=1}^{n} (\frac{N-m+l-i}{N}) \\ \mathbb{P}(C_l - C_{l-1}>n|C_{l-1}=m)&= \prod_{i=1}^{n} (1-\frac{m-l+i}{N})\end{align*} 

Lorsque $\frac{m-l+i}{N}$ est suffisamment petit, puisque $\exp x \sim 1+x$ en $0$, on aura:
\begin{align*}\mathbb{P}(C_l - C_{l-1}>n|C_{l-1}=m) &\sim \prod_{i=1}^{n} \exp-(\frac{m-l+i}{N}) \\ &\sim \exp-\sum_{i=1}^{n}\frac{m-l+i}{N}\\ \mathbb{P}(C_l - C_{l-1}>n|C_{l-1}=m) &\sim \exp-(\frac{n(m-l)}{N} + \frac{n(n+1)}{2N})\end{align*}

En particulier, lorsque $n=b\sqrt{N}$ et $m=a\sqrt{N}$, on a:$\frac{n(m-l)}{N} \rightarrow ab$ et $\frac{n(n+1)}{2N} \rightarrow \frac{b^2}{2}$ lorsque $N \rightarrow +\infty$ et finalement: $\\$
$\lim_{N\to+\infty} \mathbb{P}(C_l - C_{l-1}>b\sqrt{N}|C_{l-1}=a\sqrt{N}) = \exp-(ab + \frac{b^2}{2})$

Par suite :

$\mathbb{P}(\frac{C_l^2 - C_{l-1}^2}{2N}>y|\frac{C_{l-1}^2}{2N}=x) = \mathbb{P}(C_l - C_{l-1}>\sqrt{2N}(\sqrt{x+y}-\sqrt{x})|C_{l-1}=\sqrt{2Nx}~~) $


On peut donc appliquer le résultat précédent, avec $a=\sqrt{2x}$ et $b=\sqrt{2(x+y)} - \sqrt{2x}$. On obtient : $ab + \frac{b^2}{2} = y$ et finalement:

$\mathbb{P}(\frac{C_l^2 - C_{l-1}^2}{2N}>y|\frac{C_{l-1}^2}{2N}=x) = \exp -y$

## S1 Implémentation de l'algorithme

```python
# Importation des bibliothèques nécessaires
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as npr
```

```python
def marcheAleatoire(G, tau, l):
    
    #G est la matrice d'adjacence d'un graphe
    #Tau est la longueur des marches aléatoires
    #l est le nombre de collisions à atteindre
    
    Y      = []                   #Tableau des derniers sommets des marches aléatoires
    nb_col = 0                    #Compteur du nombre de collisions
    t      = 0                    #Compteur du nombre de marches
    
    i0 = npr.randint(0,len(G))    #Point de départ

    while nb_col < l:             #Tant que le nombre de collisions n'est pas atteint, on fait des marches aléatoires
        
        t += 1
        X = i0
        
        for i in range(tau):
            
            #Choix de changer de sommet avec probabilité 0.9
            if npr.random() < 0.9:
                
                #Construction de l'espace état à partir des arrêtes de X
                espace_etat = []
                for j in range(len(G[X])):
                    if G[X][j] != 0:
                        espace_etat.append(j)
                        
                nb_arrete = len(espace_etat)
                probas = np.full(nb_arrete,1/nb_arrete)

                #Choix du sommet suivant parmi les sommets adjacents de X avec probabilité uniforme
                X = int(npr.choice(espace_etat, p=probas, size=1)[0])
        
        #Incrémentation du compteur de collision le cas échéant
        if X in Y:
            nb_col += 1
        Y.append(X)
    
    #Quand nb_col == l, on a t == Cl
    return t*t/(2*l)   
```

## S2 Test de l'algorithme sur un hypercube

### Initialisation de l'hypercube

```python
d = 10                #Dimension de l'hypercube
N = np.power(2,d)     #Nombre de sommet de l'hypercube
H = np.zeros((N,N))   #Initialisation de la matrice d'adjacence de l'hypercube

#Les coordonnées de chaque sommet de l'hypercube forment un nombre binaire utilisé pour les indexer.
#Deux sommets sont reliés si et seulement si leur représentation binaire ne diffère que d'un seul bit.

for i in range(N):
    for j in range(N):
        if abs(bin(i & j).count("1") - bin(i | j).count("1")) == 1:
            H[i][j] = 1
```

### Initialisation des variables de simulation

```python
Tau_H     = [5, 10, 50, 100]    #Longueur d'une marche aléatoire
L_H       = [1, 10, 100]        #Nombre de collisions à atteindre
samples_H = 100                 #Nombre d'estimations de N pour chaque jeu de paramètres
```

### Calcul et sauvegarde des résulats de la simulation

```python
Resultats_H = np.zeros((len(Tau_H), len(L_H), samples_H))

for tau in range(len(Tau_H)):
    for l in range(len(L_H)):
        print(f"Calcul en cours pour tau = {Tau_H[tau]} et l = {L_H[l]}")
        for i in range(samples_H):
            Resultats_H[tau][l][i] = marcheAleatoire(H, Tau_H[tau], L_H[l])
print('Simulation terminée !')
```

### Affichage des résultats

```python
plt.figure(figsize=(16,16))
plt.suptitle("Distribution de l'estimateur $\hat{N}$ en fonction de $\\tau$ et l sur un hypercube de dimension 10", fontsize=15)

for tau in range(len(Tau_H)):
    for l in range(len(L_H)):
        
        #Moyenne de la série d'estimations
        m = np.mean(Resultats_H[tau][l])
        
        #Erreur quadratique relative
        RMS = 0
        for i in range(samples_H):
            RMS += np.power((Resultats_H[tau][l][i]/N - 1), 2)
        RMS = np.sqrt(RMS/samples_H)
        
        #Affichage
        plt.subplot(len(Tau_H),len(L_H),(len(L_H)*tau+(l+1)))
        nb_bins = 2 * int(len(Resultats_H[tau][l])**(1./3.)) 
        plt.hist(Resultats_H[tau][l], bins=nb_bins, density='true', label="N")
        plt.legend(loc='best')
        plt.title(f"$\\tau$={Tau_H[tau]} et l={L_H[l]}       Moy = {int(m)} RMS = {round(RMS,3)}")

plt.show()
```

### Interprétation des résultats

On constate qu'on obtient le meilleur résultat pour $\tau=10$ et $l=100$ avec $\mathbb{E}(\hat{N})=1024$ et une erreur quadratique relative de $0.091$. Globalement, $\tau$ et $l$ élevés donnent de meilleurs résultats. Plus $l$ est élevé, plus la répartition de $\hat{N}$ se rapproche d'une gaussienne. Pour $l=1$, la répartition est plutôt exponentielle. Ce n'est pas étonnant puisque $\hat{N}$ est alors égal à $C_l^2/2$, l'instant de la première collision, ce qui se rapproche d'une loi géométrique.

## S3 Test de l'algorithme sur un cycle

### Initialisation du cycle

```python
N = 50
G = np.zeros((N,N))

G[N-1][0] = 1
G[0][N-1] = 1
for i in range(N-1):
    G[i][i+1] = 1
    G[i+1][i] =1
print(G)
```

### Initialisation des variables de simulation

```python
Tau     = [10, 100, 1000, 10000]    #Longueur d'une marche aléatoire
L       = [1, 10, 100]              #Nombre de collisions à atteindre
samples = 100                       #Nombre d'estimations de N pour chaque jeu de paramètres
```

### Calcul et sauvegarde des résultats de la simulation

```python
Resultats = np.zeros((len(Tau), len(L), samples))

for tau in range(len(Tau)):
    for l in range(len(L)):
        print(f"Calcul en cours pour tau = {Tau[tau]} et l = {L[l]}")
        for i in range(samples):
            Resultats[tau][l][i] = marcheAleatoire(G, Tau[tau], L[l])
print('Simulation terminée !')
```

### Affichage des résultats

```python
plt.figure(figsize=(16,16))
plt.suptitle("Distribution de l'estimateur $\hat{N}$ en fonction de $\\tau$ et l sur un cycle de 50 sommets", fontsize=15)

for tau in range(len(Tau)):
    for l in range(len(L)):
        
        #Moyenne de la série d'estimations
        m = np.mean(Resultats[tau][l])
        
        #Erreur quadratique relative
        RMS = 0
        for i in range(samples):
            RMS += np.power((Resultats[tau][l][i]/N - 1), 2)
        RMS = np.sqrt(RMS/samples)
        
        #Affichage
        plt.subplot(len(Tau),len(L),(len(L)*tau+(l+1)))
        nb_bins = 2 * int(len(Resultats[tau][l])**(1./3.)) 
        plt.hist(Resultats[tau][l], bins=nb_bins, density='true', label="N")
        plt.legend(loc='best')
        plt.title(f"$\\tau$={Tau[tau]} et l={L[l]}       Moy = {int(m)} RMS = {round(RMS,3)}")

plt.show()
```

### Interprétation des résulats

On constate que l'estimateur donne de mauvais résultats sur le cycle, par rapport à l'hypercube. On retrouve le motif géométrique pour $l=1$ mais on constate surtout l'apparition de trous dans la distribution de $\hat{N}$, des valeurs sont donc beaucoup moins probables que d'autres. Cela pourrait être lié à une dépendance de la distribution des $Y_t$ à la parité du sommet initial $i_0$. Lorsqu'on choisit $i_0$ et $\tau$ pairs, il est beaucoup plus probable d'aboutir sur un sommet pair que sur un sommet impair (en raison de la distribution en $\frac{1}{10}$ et $\frac{9}{10}$), et cela fausse l'estimation.
